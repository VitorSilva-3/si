### Exercise 3: SelectPercentile using the "iris.csv" dataset 

```python
import numpy as np
import pandas as pd
from si.io.csv_file import read_csv
from si.data.dataset import Dataset
from si.feature_selection.select_percentile import SelectPercentile
```

```python
dataset = read_csv('../datasets/iris/iris.csv', sep=',', features=True, label=True)
selector = SelectPercentile(percentile=50)
selector.fit(dataset)
new_dataset = selector.transform(dataset)

print("\n---Feature statistics---")
if selector.F is not None and selector.p is not None:
    df_scores = pd.DataFrame({
        'Feature': dataset.features,
        'F_Score': selector.F,
        'p_Value': selector.p
    })
    
    print(df_scores.sort_values(by='F_Score', ascending=False))

print(f"\n Selected features: {new_dataset.features}")
```
